# 02 — Prompt y clasificación

Objetivo: cargar el system prompt aprobado y comprobar que el modelo devuelve el JSON esperado.

Este notebook no utiliza correos ficticios ni modifica Gmail.

```
import os
from pathlib import Path

from dotenv import load_dotenv
from openai import OpenAI

load_dotenv("../.env", override=True)

LLM_API_KEY = os.getenv("LLM_API_KEY")
LLM_BASE_URL = (
    os.getenv("LLM_BASE_URL")
    or "https://api.groq.com/openai/v1"
)
LLM_MODEL = os.getenv("LLM_MODEL")
LLM_TEMPERATURE = float(os.getenv("LLM_TEMPERATURE", 0.2))
LLM_MAX_TOKENS = int(os.getenv("LLM_MAX_TOKENS", 1200))

cliente = OpenAI(
    api_key=LLM_API_KEY,
    base_url=LLM_BASE_URL,
)

print("Cliente preparado:", LLM_MODEL)
```

## 1. Cargar el system prompt

```
SYSTEM_PROMPT = Path(
    "../prompts/system_prompt_secretario.md"
).read_text(encoding="utf-8")

print(SYSTEM_PROMPT[:800])
```

## 2. Comprobar el formato de respuesta

No enviamos un correo real todavía. Solo pedimos al modelo que demuestre que entiende el formato JSON.

```
respuesta = cliente.chat.completions.create(
    model=LLM_MODEL,
    messages=[
        {
            "role": "system",
            "content": SYSTEM_PROMPT,
        },
        {
            "role": "user",
            "content": (
                "Devuelve un ejemplo vacío del objeto JSON "
                "para un correo recibido. No inventes datos."
            ),
        },
    ],
    temperature=0,
    max_tokens=LLM_MAX_TOKENS,
)

print(respuesta.choices[0].message.content)
```
