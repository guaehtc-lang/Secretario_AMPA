# 01 — Configuración y primera llamada al LLM

Objetivo: cargar `.env`, crear el cliente de Groq y realizar una primera llamada.

## 1. Instalar las librerías

Descomenta la línea si todavía no están instaladas.

```
# !pip install -q openai python-dotenv
```

## 2. Cargar la configuración

```
import os
from dotenv import load_dotenv

load_dotenv("../.env", override=True)

LLM_PROVIDER = os.getenv("LLM_PROVIDER")
LLM_API_KEY = os.getenv("LLM_API_KEY")
LLM_BASE_URL = os.getenv("LLM_BASE_URL")
LLM_MODEL = os.getenv("LLM_MODEL")
LLM_TEMPERATURE = float(os.getenv("LLM_TEMPERATURE", 0.2))
LLM_MAX_TOKENS = int(os.getenv("LLM_MAX_TOKENS", 1200))

print("Proveedor:", LLM_PROVIDER)
print("Modelo:", LLM_MODEL)
print("Temperatura:", LLM_TEMPERATURE)
print("Máximo de tokens:", LLM_MAX_TOKENS)
print("API key cargada:", "Sí" if LLM_API_KEY else "No")
```

## 3. Crear el cliente de Groq

```
from openai import OpenAI

BASE_URL = (
    LLM_BASE_URL
    or "https://api.groq.com/openai/v1"
)

cliente = OpenAI(
    api_key=LLM_API_KEY,
    base_url=BASE_URL
)

print(
    f"Cliente listo | Proveedor: Groq | Modelo: {LLM_MODEL}"
)
```

## 4. Primera llamada

```
respuesta = cliente.chat.completions.create(
    model=LLM_MODEL,
    messages=[
        {
            "role": "user",
            "content": (
                "Explica en una frase cuál es la función "
                "principal del secretario de una AMPA."
            )
        }
    ],
    temperature=LLM_TEMPERATURE,
    max_tokens=LLM_MAX_TOKENS
)

print(respuesta.choices[0].message.content)
```

## 5. Consumo de tokens

```
uso = respuesta.usage

print("Tokens de entrada:", uso.prompt_tokens)
print("Tokens de salida:", uso.completion_tokens)
print("Tokens totales:", uso.total_tokens)
```
